## Exploratory Data Analysis (EDA)

### Merging Datasets


````python
# 1. Menggabungkan Data untuk Analisis Kategori Produk
merged_sales = orderItems_df.merge(
    products_df,
    on='product_id',
    how='left'
).merge(
    productCategory_df,
    on='product_category_name',
    how='left'
)

# 2. Menghitung Keterlambatan Pengiriman

orders_df['order_estimated_delivery_date'] = pd.to_datetime(orders_df['order_estimated_delivery_date'])
orders_df['order_delivered_customer_date'] = pd.to_datetime(orders_df['order_delivered_customer_date'])

orders_df['delivery_delay'] = (
    orders_df['order_delivered_customer_date'] - orders_df['order_estimated_delivery_date']
).dt.days

orders_df['delivery_delay'] = orders_df['delivery_delay'].apply(lambda x: max(x, 0))  # Menghindari nilai negatif
````

### Pertanyaan 1: Kategori Produk Paling Banyak Terjual

````python
# Hitung total penjualan per kategori
category_sales = merged_sales.groupby('product_category_name_english')['product_id'].count().reset_index()
category_sales.columns = ['category', 'total_sales']

# Tambahkan kolom persentase
category_sales['percentage'] = (category_sales['total_sales'] / category_sales['total_sales'].sum()) * 100

# Urutkan dari terbesar ke terkecil
category_sales = category_sales.sort_values('total_sales', ascending=False).reset_index(drop=True)

````

### Pertanyaan 2: Pengaruh Keterlambatan Pengiriman terhadap Review Score

````python
# Gabungkan data orders dan reviews
merged_reviews = orders_df.merge(
    orderReviews_df,
    on='order_id',
    how='inner'
)

# Hitung rata-rata delay per review score
delay_by_score = merged_reviews.groupby('review_score')['delivery_delay'].mean().reset_index()

# Analisis korelasi
correlation = merged_reviews[['delivery_delay', 'review_score']].corr().iloc[0, 1]
print(f"Korelasi antara keterlambatan dan review score: {correlation:.2f}")
````

**Insight:**
Untuk pertanyaan 1 dilakukan:
- Menggabungkan order_items dengan products berdasarkan product_id
- Hitung jumlah produk terjual per kategori

Untuk pertanyaan 2 dilakukan:
- Konversi ke datetime
- Hitung keterlambatan pengiriman (dalam hari)
- Gabungkan dengan review berdasarkan order_id

## Visualization & Explanatory Analysis

### Pertanyaan 1:

````python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=category_sales.head(10),
    x='total_sales',
    y='category',
    palette='viridis'
)
plt.title('Top 10 Kategori Produk Berdasarkan Jumlah Penjualan')
plt.xlabel('Total Penjualan')
plt.ylabel('Kategori Produk')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Tambahkan label persentase
for i, row in category_sales.head(10).iterrows():
    plt.text(
        x=row['total_sales'] + 50,
        y=i,
        s=f"{row['percentage']:.1f}%",
        va='center'
    )

plt.show()
````

### Pertanyaan 2:

````python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=delay_by_score,
    x='review_score',
    y='delivery_delay',
    palette='coolwarm'
)
plt.title('Rata-Rata Keterlambatan Pengiriman Berdasarkan Review Score')
plt.xlabel('Review Score')
plt.ylabel('Rata-Rata Keterlambatan (Hari)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Analisis korelasi
correlation = merged_reviews[['delivery_delay', 'review_score']].corr().iloc[0, 1]
print(f"Korelasi antara keterlambatan dan review score: {correlation:.2f}")
````

**Insight:**
Kategori dengan penjualan tertinggi:
1. bed_bath_table
2. health_beauty
3. sport_leisure

Dari bar plot yang telah dibuat, kita bisa mengamati:

- Jika keterlambatan meningkat, maka review score cenderung menurun.
- Review dengan skor 5 memiliki keterlambatan yang lebih kecil dibanding review dengan skor lebih rendah.
- Pelanggan yang mengalami keterlambatan ekstrem lebih cenderung memberikan rating rendah (1 atau 2).
